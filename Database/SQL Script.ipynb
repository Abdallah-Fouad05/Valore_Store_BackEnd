# [dbo].[sp_CalculateOrderTotal]

```sql
/****** Object:  StoredProcedure [dbo].[sp_CalculateOrderTotal]    Script Date: 1/21/2026 3:17:29 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_CalculateOrderTotal]
    @OrderID INT
AS
BEGIN
    DECLARE @TotalAmount DECIMAL(18,2);

    SELECT @TotalAmount = SUM(Quantity * OrderItems.Price)
    FROM OrderItems
    WHERE OrderID = @OrderID;

    UPDATE Orders
    SET TotalAmount = ISNULL(@TotalAmount, 0)
    WHERE OrderID = @OrderID;
END
GO
```

# [dbo].[sp_CreateCartItem]

```sql
/****** Object:  StoredProcedure [dbo].[sp_CreateCartItem]    Script Date: 1/21/2026 3:17:29 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_CreateCartItem]
    @UserID INT,
    @ProductID INT,
    @Quantity INT,
    @CartID INT OUTPUT,
    @CartItemID INT OUTPUT
AS
BEGIN
    SET NOCOUNT ON;

    SELECT TOP 1 @CartID = CartID
    FROM Carts
    WHERE UserID = @UserID
    ORDER BY CreatedAt DESC;

    IF @CartID IS NULL
    BEGIN
        INSERT INTO Carts(UserID, CreatedAt, UpdatedAt)
        VALUES (@UserID, SYSDATETIME(), SYSDATETIME());

        SET @CartID = SCOPE_IDENTITY();
    END

    IF EXISTS (SELECT 1 FROM CartItems WHERE CartID = @CartID AND ProductID = @ProductID)
    BEGIN
        SET @CartItemID = -1; 
    END
    ELSE
    BEGIN
        INSERT INTO CartItems(CartID, ProductID, Quantity, CreatedAt, UpdatedAt)
        VALUES (@CartID, @ProductID, @Quantity, SYSDATETIME(), SYSDATETIME());

        SET @CartItemID = SCOPE_IDENTITY();
    END
END
GO
```

# [dbo].[sp_SignUp]

```sql
/****** Object:  StoredProcedure [dbo].[sp_SignUp]    Script Date: 1/21/2026 3:17:29 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_SignUp]
    @email NVARCHAR(MAX),
    @fullName NVARCHAR(MAX),
    @password NVARCHAR(128),
    @createdat DATETIME
AS
BEGIN
    SET NOCOUNT ON;

    INSERT INTO dbo.Users
    (
        UserName,
        Email,
        [Password],
        ImageURL,
        CreatedAt,
        UpdatedAt,
        IsAdmin
    )
    VALUES
    (
        @fullName,
        @email,
        @password,
        NULL,
        @createdat,
        @createdat,
        0
    );

    -- 🔁 Return UserID + Email
    SELECT 
        SCOPE_IDENTITY() AS UserID,
        @email AS Email;
END
GO
```

# [dbo].[sp_UpdateUser]

```sql
/****** Object:  StoredProcedure [dbo].[sp_UpdateUser]    Script Date: 1/21/2026 3:17:29 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_UpdateUser]
    @UserID     INT,
    @UserName   NVARCHAR(200),
    @Email      NVARCHAR(200),
    @Password   NVARCHAR(128) = NULL,
    @ImageURL   NVARCHAR(MAX),
    @IsAdmin    BIT
AS
BEGIN
    SET NOCOUNT ON;

    UPDATE Users
    SET
        UserName  = @UserName,
        Email     = @Email,
        ImageURL  = @ImageURL,
        IsAdmin   = @IsAdmin,
        UpdatedAt = SYSDATETIME(),
        [Password] =
            CASE
                WHEN @Password IS NULL THEN [Password]
                ELSE @Password
            END
    WHERE UserID = @UserID;

	select @@ROWCOUNT As rwoUpdated;
END
GO
```
